````python
# ── Definir todas las funciones core (igual que project.py) ──────────────────
HISTORY_FILE   = "score_history.csv"
QUESTIONS_FILE = "questions.json"

def load_questions(filepath=QUESTIONS_FILE):
    if not os.path.exists(filepath):
        raise FileNotFoundError(f"Questions file not found: {filepath}")
    with open(filepath, encoding="utf-8") as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError as exc:
            raise ValueError(f"Invalid JSON in {filepath}: {exc}") from exc
    if not isinstance(data, list) or not data:
        raise ValueError("Questions file must contain a non-empty JSON array.")
    return data

def calculate_score(base_points, streak, time_bonus=0):
    if base_points < 0:
        raise ValueError("base_points must be non-negative.")
    if streak < 0:
        raise ValueError("streak must be non-negative.")
    multiplier = 1.0 + 0.5 * (streak // 3)
    multiplier = min(multiplier, 4.0)
    return int(base_points * multiplier) + time_bonus

def filter_questions(questions, categories=None, difficulty=None):
    valid_difficulties = {None, "easy", "medium", "hard"}
    if difficulty not in valid_difficulties:
        raise ValueError(
            f"difficulty must be one of {valid_difficulties}, got '{difficulty}'."
        )
    result = questions
    if categories:
        result = [q for q in result if q["category"] in categories]
    if difficulty:
        result = [q for q in result if q["difficulty"] == difficulty]
    return result

def save_score(player, score, total, category, difficulty, filepath=HISTORY_FILE):
    file_exists = os.path.exists(filepath)
    with open(filepath, "a", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(
            f, fieldnames=["timestamp","player","score","questions","category","difficulty"]
        )
        if not file_exists:
            writer.writeheader()
        writer.writerow({
            "timestamp":  datetime.now().strftime("%Y-%m-%d %H:%M"),
            "player":     player,
            "score":      score,
            "questions":  total,
            "category":   category,
            "difficulty": difficulty,
        })

def load_history(filepath=HISTORY_FILE):
    if not os.path.exists(filepath):
        return []
    with open(filepath, encoding="utf-8") as f:
        reader = csv.DictReader(f)
        return list(reader)

def format_leaderboard(history, top_n=5):
    if not history:
        return "No hay partidas registradas todavía."
    sorted_history = sorted(history, key=lambda r: int(r["score"]), reverse=True)
    top = sorted_history[:top_n]
    medals = ["🥇","🥈","🥉"] + ["  " for _ in range(top_n)]
    lines = [f"\n{'─'*52}", f"  {'LEADERBOARD':^48}", f"{'─'*52}"]
    for i, row in enumerate(top):
        lines.append(
            f"  {medals[i]}  {row['player']:<16} {int(row['score']):>5} pts"
            f"   {row['timestamp']}"
        )
    lines.append(f"{'─'*52}")
    return "\n".join(lines)

print("✅ Funciones core cargadas")
````

### Grupo 1 — `load_questions` (6 tests)

````python
%%ipytest -v

import pytest, json as _json, os as _os

SAMPLE_QUESTIONS = [
    {"category":"Química","difficulty":"easy","question":"¿Símbolo del oro?",
     "options":["Au","Ag","Fe","Cu"],"answer":"Au","explanation":"Aurum en latín."},
    {"category":"Física","difficulty":"medium","question":"¿Velocidad de la luz?",
     "options":["3×10⁸ m/s","3×10⁶ m/s","3×10⁴ m/s","3×10¹⁰ m/s"],"answer":"3×10⁸ m/s","explanation":"c ≈ 3×10⁸ m/s."},
    {"category":"Química","difficulty":"hard","question":"Variable fundamental del DFT?",
     "options":["Función de onda","Densidad electrónica","Energía","Potencial"],"answer":"Densidad electrónica","explanation":"Teoremas de Hohenberg-Kohn."},
    {"category":"Astronomía","difficulty":"easy","question":"¿Planetas en el Sistema Solar?",
     "options":["7","8","9","10"],"answer":"8","explanation":"Plutón es planeta enano desde 2006."},
]

@pytest.fixture
def sample_json(tmp_path):
    p = tmp_path / "questions.json"
    p.write_text(_json.dumps(SAMPLE_QUESTIONS), encoding="utf-8")
    return str(p)

def test_load_questions_returns_list(sample_json):
    assert isinstance(load_questions(sample_json), list)

def test_load_questions_correct_count(sample_json):
    assert len(load_questions(sample_json)) == len(SAMPLE_QUESTIONS)

def test_load_questions_has_required_keys(sample_json):
    required = {"category","difficulty","question","options","answer","explanation"}
    for q in load_questions(sample_json):
        assert required.issubset(q.keys())

def test_load_questions_file_not_found():
    with pytest.raises(FileNotFoundError):
        load_questions("/nonexistent/path/questions.json")

def test_load_questions_invalid_json(tmp_path):
    bad = tmp_path / "bad.json"
    bad.write_text("NOT VALID JSON", encoding="utf-8")
    with pytest.raises(ValueError, match="Invalid JSON"):
        load_questions(str(bad))

def test_load_questions_empty_array(tmp_path):
    empty = tmp_path / "empty.json"
    empty.write_text("[]", encoding="utf-8")
    with pytest.raises(ValueError, match="non-empty"):
        load_questions(str(empty))
````

### Grupo 6 — `format_leaderboard` (5 tests)

````python
%%ipytest -v

import pytest

def test_format_leaderboard_empty():
    assert "No hay partidas" in format_leaderboard([])

def test_format_leaderboard_contains_player_name():
    h = [{"player":"Yoh","score":"42","timestamp":"2025-01-01 10:00"}]
    assert "Yoh" in format_leaderboard(h)

def test_format_leaderboard_sorted_descending():
    h = [
        {"player":"PlayerLow",  "score":"10","timestamp":"2025-01-01 10:00"},
        {"player":"PlayerHigh", "score":"50","timestamp":"2025-01-02 10:00"},
        {"player":"PlayerMid",  "score":"30","timestamp":"2025-01-03 10:00"},
    ]
    result = format_leaderboard(h)
    assert result.index("PlayerHigh") < result.index("PlayerMid") < result.index("PlayerLow")

def test_format_leaderboard_respects_top_n():
    h = [{"player":f"P{i}","score":str(i*10),"timestamp":"2025-01-01 10:00"} for i in range(10)]
    assert "P0" not in format_leaderboard(h, top_n=3)

def test_format_leaderboard_returns_string():
    assert isinstance(format_leaderboard([{"player":"X","score":"5","timestamp":"2025-01-01 10:00"}]), str)
````
